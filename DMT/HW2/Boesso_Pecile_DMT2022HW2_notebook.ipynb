### 1.1.4
Create the adjacency matrix for the graph, assigning to each character a node identifier equal to the index that the character name has in ascending alphabetical order (remember that the first element of a list in Python has index 0).

```python
#YOUR CODE STARTS HERE#
# we create the adjacenct matrix following the ascending alphabetical order
adj = np.zeros((119,119))
for row in db.itertuples():
  name1 = row.character_1
  name2 = row.character_2
  adj[names.index(name1),names.index(name2)]+= 1
  adj[names.index(name2),names.index(name1)]+= 1










#YOUR CODE ENDS HERE#
#THIS IS LINE 20#
```

### 1.1.5
Compute the PageRank vector for the obtained graph using a damping factor of 0.8.

```python
#YOUR CODE STARTS HERE#
# we compute the pageRank vector using damping factor equal to 0.8
pagerank = PageRank(damping_factor=0.8, solver="piteration", n_iter=1000, tol=10**-6)
pagerank_vector = pagerank.fit_transform(adj)














#YOUR CODE ENDS HERE#
#THIS IS LINE 20#
```

### 1.1.6
Compute the Topic-specific PageRank vector for the obtained graph using a damping factor of 0.9, by considering as topic the Weasley family: a character belongs to the topic if its name contains the string `Weasley`.

```python
#YOUR CODE STARTS HERE#
# we compute the PageRank
pagerank = PageRank(damping_factor=0.9, solver="piteration", n_iter=1000, tol=10 ** -6)
count = 0
index = []
# we find all the characters with "Weasley"
for i in range(len(names)):
  if "Weasley" in names[i]:
    count+=1
    index.append(i)
prob = 1 / count 
diz = {i: prob for i in index} 
# we compute the topic specific PageRank vector
pagerank_vector = pagerank.fit_transform(adj,seeds=diz)




#YOUR CODE ENDS HERE#
#THIS IS LINE 20#
```

Rank all the nodes in the graph according to the PageRank score of each node: from the most important to the least important. Print this rank.

Rank all the nodes in the graph according to the inDegree score of each node: from the most important to the least important. Print this rank.

On another line, print the identifier of the node with the highest inDegree score and the identifier of the node with the highest PageRank score.

```python
#YOUR CODE STARTS HERE#
adj = np.zeros((12,12))
cords = [[1,0],[2,0],[3,0],[4,1],[5,1],[6,1],[7,1],[8,2],[9,2],[10,2],[11,3]]
for y,x in cords :
  adj[y,x]=1

pagerank = PageRank(damping_factor=0.8, solver="piteration", n_iter=1000, tol=10**-6)
pagerank_vector = pagerank.fit_transform(adj)

ranks = []
for index,score in enumerate(pagerank_vector):
  ranks.append(score)
  print( "the  node {} has score {} ".format(index,score))
ranks= np.array(ranks)
print()

indegrees=[]
for index in range(12):
  indegree = np.sum(np.transpose(adj)[index])
  indegrees.append(indegree)
  print( "the  node {} has in-degree {} ".format(index,indegree))
indegrees= np.array(indegrees)
print()
print("the identifier of the node with the highest inDegree score is {} and the identifier of the node with the highest PageRank score is {}".format(np.argmax(indegrees),np.argmax(ranks)))






#YOUR CODE ENDS HERE#
#THIS IS LINE 30#
```

### 2.2.1
**Question 1:**
The company A-DMT wants to check **23** possible configurations for the `SVD` recommendation algorithm. The number of parameters specified for the various configurations must be at least 2 and no more than 5.

The company wants to obtain the best configuration among all the 23 configurations, considering the Root Mean Squared Error metric calculated on a cross-validation of **3** folds.

1. Define and print the configurations' dictionary that will be used for the parameter optimization.
2. Find an `SVD` model configuration that offers the best possible performance within the constraints provided by the company. Print this configuration.

The resulting solution must outperform the default `SVD` configuration according to the Mean Absolute Error metric.


```python
#YOUR CODE STARTS HERE#
kf = KFold(n_splits=3, random_state=42)
# we set the parameters
grid_of_parameters = {
   'n_factors': [100,20, 40, 120],  # The MAXIMUM number of factors.
   'n_epochs': [ 20,25,30],  # The  number of epochs.
   'lr_all': [0.0025,0.0002]    
}
print( "for the parameter optimization we are going to use the following parameters:")
print( "the possible numbers of factors are: {}\nthe possible numebers of epochs are: {}\nthe possible values of learning rate are:{}".format(grid_of_parameters['n_factors'],grid_of_parameters['n_epochs'],grid_of_parameters['lr_all']))
rs = RandomizedSearchCV(surp.matrix_factorization.SVD, param_distributions=grid_of_parameters, n_iter=23, measures=['mae'], cv=kf, n_jobs=2,
                       joblib_verbose=100)
# we get the metrics
rs.fit(data)
print("THE BEST_MAE: " + str(rs.best_score['mae']))
print("BEST_PARAMETERS FOR THE CHOSEN CONFIGURATION IS: ")
print(rs.best_params['mae'])
DefaultFactors = 100
DefaultEpochs = 20
DefaultLR = 0.005
# we compare the obtained best metric with the one from a default configuration
print("These are the metrics we get with the default configurations:")
current_algo = surp.matrix_factorization.SVD(n_factors=DefaultFactors,
                  n_epochs=DefaultEpochs,
                  lr_all=DefaultLR,
                  verbose=False)

res= cross_validate(current_algo, data, measures=['mae'], cv=kf, verbose=True)
#YOUR CODE ENDS HERE#
#THIS IS LINE 30#
```

```python
#REMOVE_CELL#
import nbformat as nbf
from glob import glob
notebooks = glob("/content/**/*.ipynb", recursive=True)

text_search_dict = ["----------YOUR TEXT STARTS HERE----------","#YOUR CODE STARTS HERE#"]

for ipath in notebooks:
    print(ipath)
    ntbk = nbf.read(ipath, nbf.NO_CONVERT)

    for cell in ntbk.cells:
        cell_type = cell['cell_type'] == "code"
      
        cell_tags = cell.get('metadata', {}).get('tags', [])
        cell_tags = set(cell_tags)

        if cell_type == 0: #markdown
          if (cell['source'][0]!="#" and text_search_dict[cell_type] not in cell['source']) or "#REMOVE_CELL#" in cell['source']:
            cell_tags.add("remove_cell")

        else: #code
          if text_search_dict[cell_type] not in cell['source'] or "#REMOVE_CELL#" in cell['source']:
            cell_tags.add("remove_cell")
          elif "#REMOVE_OUTPUT#" in cell['source']:
            cell_tags.add("remove_output")
        
        if len(cell_tags) > 0:
            cell['metadata']['tags'] = list(cell_tags)

    nbf.write(ntbk, ipath)
```
