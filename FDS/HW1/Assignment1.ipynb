##### 1-D Gaussian Filter
Implement a method which computes the values of a 1-D Gaussian *Gx* for a given standard deviation $\sigma$ :
\begin{equation}
G=\frac{1}{\sqrt{2\pi}\sigma}\exp(-\frac{x^2}{2\sigma^2})
\end{equation}

The method should also return a vector *x* of values on which the Gaussian filter is defined: integer values on the interval $\left[-3\sigma,3\sigma\right]$.

This should be the output with $\sigma=3$.


![](https://drive.google.com/uc?export=view&id=1Q3WIdozsroKzinnlyZhgZHFMdYj7KQrw)


```python
def gauss(sigma):
    """
    Gauss function taking as argument the standard deviation sigma
    The filter should be defined for all integer values x in the range [-3sigma,3sigma]
    The function should return the Gaussian values Gx computed at the indexes x
    """
    # create empty lists for Gx (aka the kernel) and x values
    Gx, x = [], []
    # for each value in the range, store it in the x list 
    # then calculate the Gx and append it to the Gx list
    for i in range(-3*sigma, (3*sigma + 1)):
        Gx.append(1/(math.sqrt(2*math.pi)*sigma)*math.exp(-(i**2)/(2*sigma**2))) #Gaussian values
        x.append(i) 
    return Gx, x
```

##### 1-D Box Filter

Implement a method which computes the values of a 1-D Box *Bx* for a given standard deviation $\sigma$ :
\begin{equation}
B= [\frac{1}{n},\frac{1}{n},\dots,\frac{1}{n}]\in \mathbb{R}^n 
\end{equation}

The method should also return a vector *x* of values on which the **Box** filter is defined: integer values on the interval $\left[-3\sigma,3\sigma\right]$.

An example of Box Filter:

![](https://drive.google.com/uc?export=view&id=1UBpSRYNkZnK2UGmGsMSnrmzwrPyMt6Fz)

```python
def box(filter_size=3,show_verbose=False):
    """
    box function taking as argument the filter size.
    The filter should be defined for all integer values and centered at zero
    The function should return the Box values Bx computed at the indexes x
    """
    Bx, x = [], []
    # create a linspace of size filter_size 
    a = np.linspace((-filter_size/2), (filter_size/2), filter_size) 
    # for each value in the linspace, store it in the x list
    #then calculate the Bx and store it in the Bx list
    for i in a:
        x.append(i) #indexes
        Bx.append(1/filter_size) #Box values

    return Bx,x
```

##### Gauss derivative 1-D
Implement a function gaussdx for creating a Gaussian derivative filter in 1D:
$$
\frac{d}{dx}G =-\frac{1}{\sqrt{2\pi}\sigma^3}x\exp({-\frac{x^2}{2\sigma^2}})
$$
As in 1.a-1.b, the function should also return a vector x of values on which the Gaussian derivative filter is
defined: integer values on the interval $[3\sigma; 3\sigma]$.

An example of Gauss derivative 1-D:

![](https://drive.google.com/uc?export=view&id=1QZmVMxJiOARKe5MoqzZ7SSSjFikoY2ty)

```python
def gaussdx(sigma):
    # create empty lists for Dx (aka the kernel) and x values
    Dx, x = [], []
    
    # for each value in the range, store it in the x list 
    # then calculate the derivative in dx (aka Dx) and append it to its list
    for i in range(-3*sigma, 3*sigma):
        exp = math.exp(-(i**2)/(2*(sigma**2)))
        Dx.append(-(1/(math.sqrt(2*math.pi)*(sigma**3)))*i*exp) #Gauss derivative values
        x.append(i) 

    return Dx, x
```

The function should take an image as an input
and return the result of convolution of this image with 2D Gaussian kernel. You can take advantage of the Pythons convolve2D function if you dont want to

```python
import scipy.signal
```

```python
def gaussfiltering(img, sigma):
    """
    Implement a 2D Gaussian filter, leveraging the previous gauss.
    Implement the filter from scratch or leverage the convolve or convolve2D methods (scipy.signal)
    Leverage the separability of Gaussian filtering
    Input: image, sigma (standard deviation)
    Output: smoothed image
    """
    # get the gaussian kernel Gx and turn it in an np.array
    Gx, x = gauss(sigma) 
    Gx = np.asarray(Gx)
    smooth_img = []
    # smooth the image convolving each row with the gaussian kernel
    # iterate at first over the rows <-- x axis
    for row in img:
        new_row = scipy.signal.convolve(row, Gx, mode='same')/sum(Gx)
        smooth_img.append(new_row)
    # then iterate over the cols, after transposing the img <-- y axis
    img = np.transpose(smooth_img)
    smooth_img=[]
    for col in img:
        new_row = scipy.signal.convolve(col, Gx, mode='same')/sum(Gx)
        smooth_img.append(new_row)
    # transpose again to get the right img
    smooth_img = np.transpose(smooth_img)
    return np.array(smooth_img)
```

##### 2-D Box filter
The function should take an image as an input
and return the result of convolution of this image with 2D Box kernel. 

See figure below for illustration of Box filtering.

![Original Image](https://drive.google.com/uc?export=view&id=1IMrJUuTouVHmaiiyWB6H1QbKLzYTg9qC)



```python
def boxfiltering(img, filter_size):
    """
    Implement a 2D Box filter, leveraging the previous box.
    Leverage the separability of Box filtering
    Input: image, filter_size
    Output: smoothed image
    """
    # get the box kernel Gx and turn it in an np.array
    Bx, x = box(filter_size)
    Bx = np.array(Bx)
    # transpose to get the values for the y axis
    By = np.transpose(Bx)
    # compute the outer product, leveraging the separability of Box filter
    m = np.outer(Bx, By) 
    # leveraging the scipy function convolve2d, compute the convolved img
    smooth_img = scipy.signal.convolve2d(img, m, boundary='symm', mode='same')

    return np.array(smooth_img)
```

See figure below for an illustration of this custom filter.

![](https://drive.google.com/uc?export=view&id=1N1wEkhqOuAX3TusQNNxN9KPOXb9nRsFn)



```python
# convolve centering the kernel to the pixel to convolve
def my_convolve(kernel, row, col, h):
    pix_av = 0
    for i in range(-h, h):
        for j in range(-h, h):
            pix_av += img[row - i][col - i] * kernel[i + h][j + h]
    return pix_av

def customfiltering(img):
    """
    Implement a 2D Custom filter
    Implement the filter from scratch 
    Input: image
    Output: smoothed image
    """
    filtered_img = []
    # get the custom kernel 
    kernel = custom()
    # compute the indexes to iterate on the img
    # pay attention to exclude the borders not to make them disappear 
    # divide the kernel dims to get the right distances between its centre and preserve the borders
    h = len(kernel)//2
    h_col = len(kernel[1])//2
    # iterate over the img axis, convolving each pixel and create the filtered img
    for row in range(h, len(img)- h):
        f_img_row = []
        for col in range(h, len(img[1])- h_col):
            val = my_convolve(kernel, row, col, h)
            f_img_row.append(val)
        filtered_img.append(f_img_row)
    return np.array(filtered_img)
```

Implement a Gaussian Pyramid: Write a function for downscaling (you can use the function [resize from cv2](https://docs.opencv.org/master/dd/d52/tutorial_js_geometric_transformations.html) library) and use the Gaussian filter to blur it. The Pyramid must consist of three steps:

* Resize (Downscale) the smoothed image (obtained from previous step) and apply Gaussian filter again.
* Resize (Downscale) the smoothed image (obtained from previous step) and apply Gaussian filter again.

See figure below for an illlustration of Gaussian Pyramid.
![](https://drive.google.com/uc?export=view&id=1JKeyNozWgoBWHqrUwj4SEF6SJEnO9uMt)


```python
def downscale(img, factor):
    '''
    This function should return the given image resized by the factor.
    Input: img: the image to resize
    factor: the factor you want use to downscale
    Output: resized: the resized image
    '''
    # resize the img by the given factor
    width = len(img)-int(len(img)*factor)
    height = len(img[1])-int(len(img[1])*factor)
    resized_img = cv2.resize(img,(height,width))
    return resized_img
```

```python
def GaussianPyramid(img, steps=3, factor=0.5, sigma=4):
    '''
    This function implements the Gaussian Pyramid and shows the results.
    Leverage the "downscale" function.

    Inputs:
    img: the image
    steps: number of steps to use in performing the Pyramid
    factor: the scaling factor to resize the image at each step
    sigma: the Gaussian filter parameter

    Output:
    --
    '''
    filtered_img = img
    # create a gaussian pyramid of step steps
    for step in range(steps): 
        # if it's the first step, there's no need to resize the img
        # just filter the img
        if step == 0:
            resized_img = img
            filtered_img = gaussfiltering(img, sigma)   
        # otherwise, resize it and filter it
        else:
            resized_img = downscale(filtered_img, factor)
            filtered_img = gaussfiltering(resized_img, sigma)
        # plot the results
        fig = plt.gcf()
        DPI = fig.get_dpi()/2
        plt.rcParams["figure.figsize"] = [len(resized_img)/DPI,len(resized_img[0]/DPI)]
        plt.figure()
        plt.subplot(1,2,1)
        plt.imshow(resized_img, cmap='gray', vmin=0, vmax=255)
        plt.subplot(1,2,2)
        plt.imshow(filtered_img, cmap='gray', vmin=0, vmax=255)
```

- Image A filter -> **Gaussian Filter**:

The Gaussian filter is used to blur both grayscale and coloured images, in order to denoising it and extract useful information like edges. The Gaussian averaging in image filtering makes more 'probabilistic' sense than the other filters, because it weights nearby pixels more than distant ones.

As a linear filter, one property of the Gaussian filter is the separability, infact we can obtain a 2D convolution from two 1D convolutions. First, we convolve each row with the 1D filter and then we do the same with each column. So, we can apply together a 1D Gaussian filter horizontally and a 1D Gaussian filter vertically.

We obtain an image more naturally blurred, with less details and more focused on the objects' edges.


- Image B filter -> **Box Filter**: 

The Box Filter is a low-pass filter that smooths the image by replacing each pixel with the average of the surrounding ones, removing details and noise from images. This filter gives an equal weight to all pixels around the center without regard to the distance from the center pixel.
We can leverage the property of separability also with this linear filter, so we can get a 2D convolution from two 1D convolutions.

- Image C filter -> **Custom Filter**: 

The Custom Filter is an average linear filter where each pixels' weight is created based on our own needs.

---

We can immediately notice the differences between the three outputs: 
 - Using the Gaussian filter with =7 we obtain a very blurred image and sure enough we can't mark every brick detail but only its edges. 
 - This kind of effect is less prominent using the Box filter with size 7, as a matter of fact we have a clearer image. 
 - With the Custom filter, we obtain something different from the others because we can clearly distinguish every brick. The result is a darker image.

In general, both Box and Gaussian filter give similar results in terms of removing noise and blurring images.

But in this case, the Gaussian filter removes most of the noise from the image, in fact we can perceive only the bricks' edges. The image blurred with the Box filter keeps also some details of the bricks, not only the edges. So, if we are interested in edge detection we prefer to use the Gaussian one.

```python
def normalized_hist(img_gray, num_bins = 40, show_verbose = False):
    '''
    Compute histogram of image intensities, histogram should be normalized so that sum of all values equals 1
    assume that image intensity is between 0 and 255

    img_gray - input image in grayscale format
    num_bins - number of bins in the histogram

    '''
    # create a linspace of bins (a sample of all the 2**8 possible grey intensities)
    bins = np.linspace(0, 256, num_bins + 1) 
    # associate a counter value to each bin (initialized to 0)
    hists = np.zeros(num_bins) 
    # define a counter to count all the pixel to normalize the final result
    N = 0
    # define an interval between each bin
    i = 256/num_bins
    # iterate over the img, pixel by pixel : increment the counter value for the computed bin
    for row in img_gray:
        for pixel in row:
            N += 1
            pos = int(pixel/i)
            hists[pos] += 1
    # return the normalized histogram
    return np.array(hists)/N, bins
```

##### RGB

```python
def rgb_hist(img_color_double, num_bins = 5, show_verbose = False):
    '''
    Compute the *joint* histogram for each color channel in the image
    The histogram should be normalized so that sum of all values equals 1
    Assume that values in each channel vary between 0 and 255

    img_color - input color image
    num_bins - number of bins used to discretize each channel, total number of bins in the histogram should be num_bins^3

    E.g. hists[0,9,5] contains the number of image_color pixels such that:
      - their R values fall in bin 0
      - their G values fall in bin 9
      - their B values fall in bin 5
    '''
    # def an interval between each bin, def a pixel counter
    i, n = 256 / num_bins, 0
    # create  3D space of bins (3 samples of all the 2**8 possible r, g and b intensities)
    hists = np.zeros((num_bins, num_bins, num_bins))
    # iterate over the img, pixel by pixel
    #increment the counter value for the computed bin in 3D space
    for row in img_color_double:
        for pixel in row:
            n += 1
            r, g, b = int(pixel[0] / i), int(pixel[1] / i), int(pixel[2] / i)
            hists[r, g, b] += 1
    # return the flattened 1D histogram
    return np.ndarray.flatten(hists)/n
```

##### RG

```python
def rg_hist(img_color_double, num_bins = 5, show_verbose = False):
    '''
    Compute the *joint* histogram for the R and G color channels in the image
    The histogram should be normalized so that sum of all values equals 1
    Assume that values in each channel vary between 0 and 255

    img_color - input color image
    num_bins - number of bins used to discretize each channel, total number of bins in the histogram should be num_bins^2

    E.g. hists[0,9] contains the number of image_color pixels such that:
      - their R values fall in bin 0
      - their G values fall in bin 9
    '''
    # def an interval between each bin, def a pixel counter
    i = 256/num_bins
    # create  a linspace of bins ( a samples of all the 2**8 possible channel intensities)
    bins = np.linspace(0, 255 - i, num_bins)
    # compute the product to get a 2D space for the bins
    prod=[[x,y] for x in bins for y in bins]
    hists = np.zeros(len(prod))
    N = 0
    # iterate over the pixels and increment the value of the computed bin image
    for row in img_color_double:
        for pixel in row:
            r,g,b = pixel
            N += 1
            for x in range(len(prod)):
                if prod[x][0]<= r <prod[x][0]+i and prod[x][1]<= g <prod[x][1]+i:
                    hists[x] += 1
                    break
    hists=hists/N
    # return the normalized histogram
    return hists
```

```python
def gaussderiv(img, sigma):
    # get the kernel from the gaussian derivative
    kernel, x = gaussdx(sigma)
    # define the separated filtered img
    img_Dx, img_Dy = [], []
    
    # get values for X
    for row in img:
        new_row = scipy.signal.convolve(row, kernel, mode='same')
        img_Dx.append(new_row)
        
    # get values for Y <-- transpose to get the img Y wise   
    for col in np.transpose(img):
        new_row = scipy.signal.convolve(col, kernel, mode='same')
        img_Dy.append(new_row)
    # traspose again to set it back to X wise
    img_Dy = np.transpose(img_Dy)
    
    # cap the values
    for x in range(len(img_Dx)):
        for y in range(len(img_Dx[0])):
            if img_Dx[x][y] < -6.0: 
                img_Dx[x][y] = -6.0
            if img_Dx[x][y] > 6.0: 
                img_Dx[x][y] = 6.0
            if img_Dy[x][y] < -6.0: 
                img_Dy[x][y] = -6.0
            if img_Dy[x][y] > 6.0: 
                img_Dy[x][y] = 6.0

    return img_Dx, img_Dy
```

```python
def dxdy_hist(img_gray, num_bins = 5, show_verbose = False):
    '''
    Compute the *joint* histogram of Gaussian partial derivatives of the image in x and y direction
    Set sigma to 3.0 and cap the range of derivative values is in the range [-6, 6]
    The histogram should be normalized so that sum of all values equals 1

    img_gray - input gray value image
    num_bins - number of bins used to discretize each dimension, total number of bins in the histogram should be num_bins^2

    Note: you can use the function gaussderiv from the Filtering exercise
    '''
    # after getting the separated imgs, compute the histogram
    img_Dx, img_Dy = gaussderiv(img_gray, 3)
    # def an interval between each bin and a counter of pixel to normalize the final result
    i, n = 12/num_bins, 0
    # create a 2D bins space
    hists = np.zeros((num_bins, num_bins))
    # iterate over the pixels and invrement the computed bin image
    for x in range(len(img_Dx)):
        for y in range(len(img_Dx[0])):  #  each value in range -6 <= y <= 6
            n += 1
            x_value, y_value = img_Dx[x][y] + 6, img_Dy[x][y] + 6
            hists[int(x_value/i) - 1, int(y_value/i) - 1] += 1
    #return the flattened and normalized histogram
    return np.ndarray.flatten(hists)/n
```

#### Exercise 2.c **(1 Point)**
Implement the histogram distance functions. 

In more details, implement the function *dist_intersect*:
\begin{equation}
\bigcap(Q,V)=1-\frac{1}{2}\left(
\frac{\sum_i \min(q_i,v_i)}{\sum_i q_i} +
\frac{\sum_i \min(q_i,v_i)}{\sum_i v_i}
\right)
\end{equation}

the *dist_l2* function:
\begin{equation}
d(Q,V)=\sqrt{\sum_i (q_i-v_i)^2}
\end{equation}

and the *dist_chi2*:
\begin{equation}
\chi^2(Q,V)=\sum_i \frac{(q_i-v_i)^2}{q_i + v_i}
\end{equation}


```python
def dist_intersect(x,y):
    '''
    Compute the intersection distance between histograms x and y
    Return 1 - hist_intersection, so smaller values correspond to more similar histograms
    Check that the distance range in [0,1]
    '''
    sum_min, sum_x, sum_y = [], [], []
    # get a list of all the X and Y values, and of all the mins between them
    for i in range(len(x)):
        sum_min.append(min(x[i], y[i]))
        sum_x.append(x[i])
        sum_y.append(y[i])
    # sum the lists to get the total sum of each list
    S_min = sum(sum_min)
    S_x = sum(sum_x)
    S_y = sum(sum_y)
    # implement the intersection distance formula
    hist_intersection = (1/2)*((S_min/S_x) + (S_min/S_y))
    # warn and return nothing if the result is not compatible to theory
    if hist_intersection < 0 and hist_intersection > 1:
        print('Distance range must be in [0, 1]')
        return
    # return the computed distance
    return 1 - hist_intersection

def dist_l2(x,y):
    '''
    Compute the L2 distance between x and y histograms
    Check that the distance range in [0,sqrt(2)]
    '''
    # implement the L2 distance formula
    diff = sum([(x[i]-y[i])**2 for i in range(len(x))])
    l2_dist = math.sqrt(diff)
    # warn and return nothing if the result is not compatible to theory
    if l2_dist < 0 and l2_dist > math.sqrt(2):
        print('Distance must be in range [0, sqrt(2)]')
        return
    # return the computed distance
    return l2_dist


def dist_chi2(x,y):
    '''
    Compute chi2 distance between x and y
    Check that the distance range in [0,Inf]
    Add a minimum score to each cell of the histograms (e.g. 1) to avoid division by 0
    '''
    for i in range(len(x)):
        if x[i] < 1:
            x[i] = 1
        if y[i] < 1:
            y[i] = 1
    # implement the chi2 distance formula
    chi2 = [((x[i]-y[i])**2)/(x[i]+y[i]) for i in range(len(x))]
    chi2_dist = sum(chi2)
    # warn and return nothing if the result is not compatible to theory
    if chi2_dist < 0:
        print('Distance range must be in [0, Inf]')
        return
    # return the computed distance
    return chi2_dist
```
